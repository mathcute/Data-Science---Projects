##ATIVIDADE 3 - REGRESSAO LINEAR

```python
import numpy as np

x = np.array([9714, 3728, 6062, 8845, 8378]) #variavel independente
y = np.array([3011, 1305, 1879, 2654, 2849]) #variavel dependente

#calculo a regressao linear simples

coeficiente_angular, intercepto = np.polyfit(x, y, 1)

#coeficientes de regressao

print(f"Coeficiente Angular: {coeficiente_angular}")
print(f"Intercepto: {intercepto}")

#Criar previsoes usando a equa√ßao de regressao

y_pred = coeficiente_angular * x + intercepto
print(y_pred)

#tra√ßando dados originais e a linha de regressao

plt.scatter(x, y, label = 'Dados originais')
plt.plot(x, y_pred, color = 'black', linewidth = 2, label = 'Linha de Regressao')
plt.xlabel('Variavel Independente (X)')
plt.ylabel('Variavel dependente (Y)')
plt.legend()
plt.title('Regressao Linear Simples com Numpy')
plt.grid(True)
plt.show()
```

##REGRESSAO LINEAR 2

```python
#Dados de exemplo: numero de quartos e precos de casas

num_quartos = np.array([1, 2, 3, 4, 5, 6, 7])
preco_casas = np.array([10000, 20000, 30000, 40000, 50000, 60000, 70000])

#Calcula a regressao linear simples

coeficiente_angular, intercepto = np.polyfit(num_quartos, preco_casas, 1)

#Coeficientes da regressao

print(f"Coeficiente Angular: {coeficiente_angular}")
print(f"Intercepto: {intercepto}")

#Cria previsoes usando a equacao da regressao

y_pred = coeficiente_angular * num_quartos + intercepto
print(y_pred)

#Tra√ßando os dados originais e a liha de regressao

plt.scatter(num_quartos, preco_casas, label = 'Dados originais')
plt.plot(num_quartos, y_pred, color = 'black', linewidth = 2, label = 'Linha de Regressao')
plt.xlabel('Numero de Quartos (X)')
plt.ylabel('Preco de casas (Y)')
plt.legend()
plt.title('Regressao Linear Simples com Numpy')
plt.grid(True)
plt.show()
```
